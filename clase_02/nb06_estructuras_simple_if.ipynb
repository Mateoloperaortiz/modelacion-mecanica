# Fundamentos de programación

## Estructuras de Decisión

Las estructuras de decisión corresponden a puntos de ramificación del programa donde este puede tomar diferentes rutas a partir del resultado (Verdadero o Falso) de una condición. Estas podrían entenderse como el analogo a una pregunta en lenguaje convencional. La evaluación de la condición en estas estructuras generalmente se realiza mediante expresiones lógicas en términos de operadores lógicos (and, not, or) y de comparasión ($>, <, ==$).

### Condicioanl simple
La mas simple de las estructuras de decisión es el `condicional simple`. En este caso si la condición es Verdadera el programa ejecuta un conjunto de acciones determinadas, pero si es Falsa el programa sigue su flujo normal.

### Ejemplo

El siguiente programa lee por pantalla 2 valores enteros y los almacena en las variables $A$ y $B$. Posteriormente identifca el valor mayor y menor respectivamente.

```python
A = int(input())
B = int(input())
mayor = A
menor = B
if (A<B):
    mayor = B
    menor = A
print('mayor=', mayor)
print('menor=', menor)
```
