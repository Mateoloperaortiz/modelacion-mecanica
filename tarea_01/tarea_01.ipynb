### 1) Ensamblaje de matrices elementales en mecánica computacional (20 puntos)

En este algoritmo se desean almacenar (acumulando) los datos numericos de $Q$ matrices cuadradas $k_l$ cada una de orden $n \times n$ en una gran matriz $K_G$ de orden $M \times M$. Este proceso, conocido como de **ensamblaje**, es común en programas de Mecánica Computacional para análisis de estructuras y sólidos.

Las posiciones que ocuparán en la matriz $K_G$ cada uno de los elementos de las matrices $k_l$ se almacenan a la vez en una matriz denotada como `MIE()`. En esta última el numero de filas corresponde al numero de matrices locales a ensamblar y el numero de columnas corresponde al orden de las matrices locales. Por ejemplo una matriz `MIE` de la forma


$$ {MIE} = \left[ {\begin{array}{*{20}{c}}
0&1&{2}&{4}\\
{1}&2&5&{4}\\
4&5&{8}&{7}\\
{3}&{4}&{7}&{6}
\end{array}} \right]$$

sería valida para un modelo con 4 matrices del tipo $k_l$, cada una de orden 4.


El proceso de ensamblaje se reduce entonces a:

* localizar la fila de la matriz `MIE()` correspondinete a la matriz local a ensamblar.

* buscar la dirección de cada coeficiente en la fila correspondiente de la matriz `MIE()` y

* usar esta dirección para **ensamblar** el coeficiente en la matriz global.

Por ejemplo, suponga que una matriz $K_l$ de orden $4 \times 4$ es de la forma:

$${k^L} = \left[ {\begin{array}{*{20}{c}}
{{k_{11}}}&{{k_{12}}}&{{k_{13}}}&{{k_{14}}}\\
{{k_{21}}}&{{k_{22}}}&{{k_{23}}}&{{k_{24}}}\\
{{k_{31}}}&{{k_{32}}}&{{k_{33}}}&{{k_{34}}}\\
{{k_{41}}}&{{k_{42}}}&{{k_{43}}}&{{k_{44}}}
\end{array}} \right]$$


y que las direcciones de ensamblaje para este elemento en el arreglo `MIE()` estan dadas por:

$$\left[ {\begin{array}{*{20}{c}}
5&6&{11}&{12}
\end{array}} \right]$$

el ensamblaje de la primera fila de la matriz local $k^l$ en la matriz global $K_G$ estaría dado por:


$$\begin{array}{l}
K_{5,5}^G \leftarrow K_{5,5}^G + k_{1,1}^L\\
K_{5,6}^G \leftarrow K_{5,6}^G + k_{1,2}^L\\
K_{5,11}^G \leftarrow K_{5,11}^G + k_{1,3}^L\\
K_{5,12}^G \leftarrow K_{5,12}^G + k_{1,4}^L
\end{array}$$

## 3) Difusión de un contaminante en un rio (20 puntos)
    
Un ingeniero desea saber la variación de los niveles de contaminación producidos por el vaciado de un contaminante en una zona de un río con una extensión de $10.0$ kms. Tras probar varios modelos matemáticos se encuentra que la solución que mejor reproduce el fénomeno esta dada por la función:

$$u(x,t) = \frac{{400}}{\pi }\sum\limits_{n = 1,3,...}^\infty  {\frac{1}{n}} \sin \left( {\frac{{n\pi }}{{10}}x} \right){e^{ - 0.1125{n^2}t}}$$

donde $t$ representa el tiempo en segundos y $x$ representa la distancia a lo largo del río. Grafique la función asegurandose que la misma incluye suficientes términos  para que pueda considerarse precisa.


</div>

```python
import numpy as np
import matplotlib.pyplot as plt

x=10 # area de 10km
n=500 # 500 terminos

# se crea el array de 1 a n
terms=np.arange(1,n+1)

# se crea el array con los valores de t
t=np.linspace(0,1000,500) # 500 espacios igualmente divididos de 0 a 1000

# para cada t encontrar la variacion de contaminacion
poll_level=[] # una lista vacia
for tval in t:
    
    # vector
    poll=(400/np.pi)*np.sum((1/terms)*np.sin(terms*np.pi*x/10)*np.exp(-0.1125*n**2*tval))
    
    # guardar el sondeo a la lista poll_level
    poll_level.append(poll)
    

# grafica
plt.plot(t,poll_level,color='r')
plt.xlabel("Tiempo[segundos]")
plt.ylabel("Nivel de contaminacion")
plt.title("Variacion de los niveles de contaminacion en un area de 10km ")
plt.show()
```
